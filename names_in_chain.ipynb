```python
import requests
import re
import pypinyin as py
import numpy as np
import pandas as pd
```

# 各个包的用途概述

- 导入requests用于html的爬取
- 导入re用于正则表达式的筛选
- pypinyin库来自github：https://github.com/mozillazg/python-pinyin
- numpy用于将pypinyin生成的二维array转换为一维array
- pandas用于将numpy生成的一维array转换为list

```python
url='http://www.360doc.com/content/16/0419/10/877149_551849520.shtml'
r = requests.get(url)
r.encoding = 'utf-8'
```

- 原始数据来自360doc
- r作为获取网络内容，用UTF-8进行编码

```python
s_name = re.findall('[\u4e00-\u9fa5]{1,3}市', r.text)
s_name = list(set(s_name))
shi = []
for i in s_name:
    shi.append(i[:-1])
shi_removed = shi.remove('中国省')
shi_removed = shi.remove('国各省')
shi_removed = shi.remove('县级')
shi_removed = shi.remove('辖地级')
shi_removed = shi.remove('县个级')
shi_removed = shi.remove('全国省')

shi
```

# 结尾为市的地名的获取
- 采用正则表达式'[\u4e00-\u9fa5]{1,3}市',获取所有的市名
- 中国大陆市名多为x市、xx市或者xxx市，比如北京市、哈尔滨市、攀枝花市等，有较规律的命名方式

```python
x_name = re.findall('[\u4e00-\u9fa5]{1,3}县', r.text)
x_name = list(set(x_name))
xian = []
for i in x_name:
	xian.append(i[:-1])
xian
```

# 结尾为县的地名的获取
- 采用正则表达式'[\u4e00-\u9fa5]{1,3}县',获取所有的县名
- 中国大陆市名多为x市、xx市或者xxx市，比如河北的蓟县、安徽的全椒县、新疆的玛纳斯县等，有较规律的命名方式

```python
q_name = re.findall('[\u4e00-\u9fa5]{2,3}区', r.text)
q_name = list(set(q_name))
qu = []
for i in q_name:
	qu.append(i[:-1])
qu.remove('国地名')
qu.remove('各省市')
qu.remove('个市辖')
qu.remove('族自治')
qu.remove('省行政')
qu.remove('省市县')
qu.remove('个地')
qu.remove('尔自治')
qu.remove('别行政')
qu.remove('藏自治')
qu.remove('古自治')
qu.remove('万山特')
qu.remove('六枝特')

qu.append('香港')
qu.append('西藏')
qu.append('澳门')
qu.append('新疆')
qu.append('宁夏')
qu.append('广西')

qu
```

# 结尾为区的地名的获取
- 采用正则表达式'[\u4e00-\u9fa5]{1,3}区',获取所有的县名
- 中国大陆市名多为xx区或者xxx区，比如北京的朝阳区、重庆的九龙坡区等，有较规律的命名方式
- 另外在本例中，html中有较多的干扰内容，如国地名、各省市、个市辖等，需要手动一一剔除

```python
all_name = list(set(shi+xian+qu))
all_name = np.reshape(all_name,-1)
all_name
```

# 将所有的市、县、区地名集合起来
- 然后用set和list方法将其**去重**
- 然后将其转换为array，方便之后用pandas生成dataframe

```python
len(all_name)
```

# 一共提取1920个地名

```python
all_pinyin = py.pinyin(all_name)
all_pinyin = np.reshape(all_pinyin,-1)
all_pinyin
```

# 用Github上的pypinyin包
- 用py.pinyin()函数，得到所有汉字对应的拼音
- 用np.reshape()函数，将pypinyin返回的二维array，reshape成一维array，方便之后操作

```python
len(all_pinyin)
```

# 一共获取了3940个拼音

```python
word_length = []
for i in all_name:
    word_length.append(len(i))
word_length = np.array(word_length)
word_length
```

# 地名非等长的解决办法
- 有的地名是两个字，如深圳。
- 有的地名是三个字，如哈尔滨
- 通过计算每一个地名的字符串长度，实现地名的第一个字和最后一个字的拼音向对应

# 实现首字的拼音和尾字的拼音的对应，是形成押韵的关键算法

```python
len(word_length)
```

# 验证了地名长度的个数和地名的个数一样都是1920

```python
count = 0
end_position = []
for i in word_length:
    count+=i
    end_position.append(count)
end_position = np.array(end_position)
end_position
```

# 对尾拼音的定位
- **通过引入count**
- 循环叠加地名的长度，实现尾拼音的定位

```python
len(end_position)
```

```python
front_position = [m - n + 1 for m,n in zip(end_position,word_length)] 
front_position = np.array(front_position)
front_position
```

# 头拼音的定位
- 通过尾拼音的定位
- 地名的长度
- 定位出头拼音的坐标

```python
front_pinyin = []
for i in front_position:
    front_pinyin.append(all_pinyin[i - 1])
front_pinyin = np.reshape(front_pinyin,-1)
front_pinyin
```

# 头拼音列表的生成
- 有了头拼音的坐标
- 通过循环索引，便能够生成头拼音的列表

```python
end_pinyin = []
for i in end_position:
    end_pinyin.append(all_pinyin[i - 1])
end_pinyin = np.reshape(end_pinyin,-1)
end_pinyin
```

# 同理：生成尾列表

```python
df = pd.DataFrame(all_name,np.arange(len(all_name)))

df[1] = front_pinyin
df[2] = end_pinyin
df.columns = ["地名","头拼音","尾拼音"]
df
```

# 使用pandas，让数据表示更直观

```python
place_list = df['地名'].tolist()
place_list
```

# 最终分别生成三个包含地名、头拼音、尾拼音的列表，便于检索

```python
head_pinyin_list = df['头拼音'].tolist()
head_pinyin_list
```

```python
end_pinyin_list = df['尾拼音'].tolist()
end_pinyin_list
```

```python
get_name = input("请输入地名（比如北京）：")
while get_name:
    if get_name == 'QUIT':
        exit()
    else:
        if get_name in place_list:
            print("存在这个地名\n")
            if end_pinyin_list[place_list.index(get_name)] in head_pinyin_list:
                print('并且存在后继地名，后继城市有如下：\n')
                place_count = 0
                for i in place_list:
                    if end_pinyin_list[place_list.index(get_name)] == head_pinyin_list[place_list.index(i)]:
                        place_count+=1
                        print(place_count,i)
                print('选择一个地名来继续地名接龙,输入其中一个可能的地名：')
                get_name = input('哪一个地名你选择？退出请输入QUIT')
            else:
                print('但是不存在后继地名,请换一个地名重新开始。')
                get_name = input()
        else:
            print("不存在这个地名")
            print('但是不存在后继地名,请换一个地名重新开始。')
            get_name = input()

```
